# PHYTON BASICS

En este proyecto se utilizaron las librerías NumPy, Pandas y Time. NumPy se empleó para el manejo y procesamiento de arreglos numéricos, Pandas para la organización y análisis de datos en estructuras tipo DataFrame, y Time para medir el tiempo de ejecución de ciertas operaciones. Estas herramientas permiten trabajar de manera eficiente con datos y evaluar el rendimiento del código durante su ejecución.


```python
import numpy as np
import time
import pandas as pd
```

# Medición de rendimiento usando un ciclo for
En este fragmento de código se crea una lista con cien millones de números enteros y se calcula la suma de todos sus elementos utilizando un ciclo for. Para evaluar el rendimiento, se mide el tiempo de ejecución del proceso mediante la función perf_counter() de la librería Time, lo que permite analizar el costo computacional de realizar operaciones iterativas sobre grandes volúmenes de datos.


```python
lista = list(range(100_000_000))

inicio = time.perf_counter()
suma= 0
for i in lista:
  suma += i
final = time.perf_counter()
tiempo = final-inicio

print(f"tiempo de ejecucuión :{tiempo:6f} segundos")
```

# Suma de elementos usando NumPy

En este código se utiliza NumPy para generar un arreglo de mil millones de elementos y calcular la suma total mediante la función vectorizada np.sum(). A diferencia del uso de ciclos tradicionales, NumPy realiza la operación de forma optimizada a nivel interno, lo que reduce significativamente el tiempo de ejecución y demuestra la eficiencia del procesamiento vectorizado para manejar grandes volúmenes de datos.

```python
np_lista = np.arange(1000000000)

inicio = time.perf_counter()
suma_np = np.sum(np_lista)
final = time.perf_counter()

tiempo = final - inicio
print(f"Tiempo de ejecución: {tiempo:.6f} sec")
```

# Acceso a datos estructurados mediante índices
En este fragmento de código se define una base de datos sencilla utilizando una lista de listas, donde cada fila representa la información de una persona. Mediante diccionarios se asignan índices a las filas y columnas, lo que permite acceder de forma clara y organizada a un dato específico. En este caso, se obtiene el animal asociado a la persona llamada Kevin, demostrando una manera básica de manejar y consultar datos estructurados en Python.

```python
base_de_datos = [
["Oscar", 21, "Gatos"],
["Paula", 21, "Tiburones"],
["Jonatan", 24, "Hurones"],
["Kevin", 22, "Periquitos"],
["Lucero", 21, "Perros"],
["Israel", 21, "Lobos"],
["Sergio", 21, "Perros"],
["Vahir", 21, "Leones"],
["Diego", 21, "Pinguinos"],
]

columnas = {"Nombre":0, "Edad":1, "Animal":2}
filas    = {"Oscar":0, "Paula":1, "Jonatan":2, "Kevin":3, "Lucero":4, "Israel":5, "Sergio":6, "Vahir":7, "Diego":8}
print(base_de_datos[filas["Kevin"]][columnas["Animal"]])
```

# Manejo de datos con DataFrames en Pandas
En este código se utiliza la librería Pandas para convertir una base de datos almacenada en listas en un DataFrame, lo que facilita la visualización y manipulación de la información. Se asignan nombres a las columnas y un índice personalizado a las filas. Posteriormente, se realiza una consulta para obtener el animal asociado a la persona llamada Kevin, mostrando la eficiencia de Pandas para filtrar y acceder a datos de forma sencilla y legible.

```python
dF = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))
print(dF)
print(dF.Animal[dF['Nombre'] == 'Kevin'])
```
